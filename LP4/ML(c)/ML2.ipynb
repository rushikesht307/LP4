```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("emails.csv")
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#selects all rows of the DataFrame df but only the columns from index 1 to index 3000
X = df.iloc[ : , 1:3001 ]
X
```

```python
# Select all rows from the last column of the DataFrame
Y = df.iloc[:, -1]
Y
```

```python
# Split the dataset into training and testing sets
# X contains the features, Y contains the labels.
# The test_size parameter specifies that 20% of the data will be used for testing,
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Create an instance of the KNeighborsClassifier with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model using the training data (X_train) and corresponding labels (Y_train)
knn.fit(X_train, Y_train)
```

```python
# Make predictions
Y_pred_knn = knn.predict(X_test)
```

```python
# Evaluate the model
print("k-NN Accuracy:", accuracy_score(Y_test, Y_pred_knn))
print("k-NN Classification Report:\n", classification_report(Y_test, Y_pred_knn))
print("k-NN Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred_knn))
```

```python
'''
Actual -  Predicted	         Not Spam (0)	Spam (1)

Not Spam(0)	                      627	    102
Spam (1)	                      49	    257
====> 
True Negatives (TN): 627 — correctly predicted as not spam.

False Positives (FP): 102 — incorrectly predicted as spam (false alarms).

False Negatives (FN): 49 — incorrectly predicted as not spam (missed spam).

True Positives (TP): 257 — correctly predicted as spam.
 ''' 
```

```python
# Using Support  Vector Machine

# Initialize the Support Vector Machine (SVM) classifier with a linear kernel
# The linear kernel is suitable for linearly separable data and helps to create a hyperplane
svm = SVC(kernel='linear')

# Train the SVM model using the training data (X_train for features and Y_train for labels)
svm.fit(X_train, Y_train)
```

```python
# Use the trained SVM model to make predictions on the test dataset
# The predict method takes the feature set (X_test) and outputs the predicted class labels (Y_pred_svm)
Y_pred_svm = svm.predict(X_test)
```

```python
# Evaluate the model
print("SVM Accuracy:", accuracy_score(Y_test, Y_pred_svm))
print("SVM Classification Report:\n", classification_report(Y_test, Y_pred_svm))
print("SVM Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred_svm))
```
