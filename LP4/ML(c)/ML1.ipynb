```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def gradient(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent algorithm to find the local minimum of a function
def gradient_descent(starting_point, learning_rate, n_iterations):
    """
    Performs gradient descent to find the local minimum of a function.
    
    Parameters:
    starting_point (float): The initial x value to start the descent.
    learning_rate (float): The step size that controls how big a step is taken in the direction of the negative gradient.
    n_iterations (int): Number of iterations to run the algorithm.
    
    Returns:
    x (float): The final x value where the local minimum is found.
    x_values (list): A list of x values that the algorithm passes through during the descent.
    """
    
    # Initialize the starting point
    x = starting_point
    
    # Create a list to store x values at each iteration (for visualization purposes)
    x_values = [x]
    
    for i in range(n_iterations):
        
        # Compute the gradient at the current x value
        grad = gradient(x)
        
        # Update x by moving in the opposite direction of the gradient
        # This is the core of the gradient descent: x_new = x_old - learning_rate * gradient
        x = x - learning_rate * grad
        
        # Store the updated x value in the list for future plotting
        x_values.append(x)
        
        # Print the current iteration, the new x value, and the function's value at that point
        # This helps in observing how the function value decreases and x converges to the minimum
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {function(x):.4f}")
    
    # Return the final x value (the local minimum) and the list of x values traversed during descent
    return x, x_values
```

```python
# Parameters for Gradient Descent
starting_point = 2   # Start at x = 2
learning_rate = 0.1  # Learning rate
n_iterations = 20    # Number of iterations
```

```python
# Run gradient descent
x_min, x_values = gradient_descent(starting_point, learning_rate, n_iterations)
```

```python
# Plot the function and the steps of gradient descent

# Create an array of x values ranging from -10 to 5
# This will be used to plot the function over a wide range of x values
x_range = np.linspace(-10, 5, 100)

# Calculate the corresponding y values for the function y = (x + 3)^2
# This gives us the points on the curve to plot
y_range = function(x_range)

# Set the figure size for the plot (8 inches by 5 inches)
plt.figure(figsize=(8, 5))

# Plot the function y = (x + 3)^2 using the x_range and y_range values
plt.plot(x_range, y_range, label='y = (x + 3)^2')

# Plot the points from gradient descent steps on the same graph
# We convert the x_values list to a numpy array for compatibility
# Each step is marked with a red dot, showing how the algorithm moves towards the minimum
plt.scatter(x_values, function(np.array(x_values)), color='red', label='Gradient Descent Steps')

# Set the title of the plot
plt.title('Gradient Descent to Minimize (x + 3)^2')

# Label the x-axis as 'x' and the y-axis as 'y'
plt.xlabel('x')
plt.ylabel('y')

# Display the legend to differentiate between the function and the gradient descent steps
plt.legend()

# Add a grid to the plot for better readability
plt.grid(True)

# Display the plot on the screen
plt.show()
```

```python
print(f"Local minimum occurs at x = {x_min:.4f}, f(x) = {function(x_min):.4f}")
```

### What is Gradient?
=> 
A gradient is nothing but a derivative that defines the effects on outputs of the function with a little bit of variation in inputs.

### Gradient Descent Algorithm
Gradient Descent is an optimization algorithm used to minimize the value of a function 
by iteratively moving towards its local minimum. 
It is commonly used in machine learning and deep learning
for optimizing the parameters of models, such as weights in neural networks.
